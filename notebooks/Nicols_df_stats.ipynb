```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_name"
```

```python
response = requests.get(url)
response = response.content
```

```python
soup = BeautifulSoup(response, "lxml")
```
